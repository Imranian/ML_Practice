```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)
```

```python
ds = pd.read_csv('/content/drive/MyDrive/datasets/drug200.csv')
ds
```

```python
x = ds.drop(columns=['Drug','Sex', 'BP', 'Cholesterol'])
y = ds['Drug']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=125)
```

```python
# Build a Gaussian Classifier
model = GaussianNB()


# Model training
model.fit(X_train, y_train)

# Predict Output
# Use iloc to access the 7th row
predicted = model.predict([X_test.iloc[6]])

print("Actual Value:", y_test.iloc[6]) # Use iloc here as well to be consistent
print("Predicted Value:", predicted[0])
```

```python
y_pred = model.predict(X_test)
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```
